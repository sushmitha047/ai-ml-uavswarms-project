```python
# to tacle calss imbalance issue:
import torch
# Initialize counts
attack_count = 0
natural_count = 0

# Loop through each client dataset
for df in clients:
    attack_count += (df['target'] == 1).sum()  # Summing up all rows where target column is 1
    natural_count += (df['target'] == 0).sum()  # Summing up all rows where target column is 0

# Output the counts for verification
print(f"Total 'Attack' instances: {attack_count}")
print(f"Total 'Natural' instances: {natural_count}")

# Calculate total instances for normalization
total = attack_count + natural_count

# Calculate weights for each class
weight_for_attack = total / attack_count
weight_for_natural = total / natural_count

# Creating tensor for class weights
class_weights = torch.tensor([weight_for_natural, weight_for_attack], dtype=torch.float32)
```

```python
import pandas as pd
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
class ClientData(Dataset):
    def __init__(self, features, labels):
        self.features = torch.tensor(features, dtype=torch.float32)
        self.labels = torch.tensor(labels, dtype=torch.long)

    def __len__(self):
        return len(self.features)

    def __getitem__(self, index):
        return self.features[index], self.labels[index]
```

```python
from sklearn.preprocessing import StandardScaler

def prepare_data(dataframes):
    clients_loaders = []
    for df in dataframes:
        # Shuffle the dataframe
        df = shuffle(df)
        # Splitting the data into features and labels
        features = df.iloc[:, :-1].values  # Assuming the last column is the label
        labels = df.iloc[:, -1].values

        # Splitting into training, validation, and test sets with stratification
        X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.20, random_state=42, stratify=labels)
        X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.11, random_state=42, stratify=y_train)

        # Standardize the data
        scaler = StandardScaler()
        X_train = scaler.fit_transform(X_train)
        X_val = scaler.transform(X_val)
        X_test = scaler.transform(X_test)

        # Converting to PyTorch datasets and loaders
        train_dataset = ClientData(X_train, y_train)
        val_dataset = ClientData(X_val, y_val)
        test_dataset = ClientData(X_test, y_test)

        train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
        val_loader = DataLoader(val_dataset, batch_size=64, shuffle=False)
        test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)

        clients_loaders.append((train_loader, val_loader, test_loader))

    return clients_loaders
```

```python
class NeuralNet(nn.Module):
    def __init__(self):
        super(NeuralNet, self).__init__()
        self.layer1 = nn.Linear(128, 64)  # Assuming each data sample has 128 features
        self.bn1 = nn.BatchNorm1d(64)
        self.act1 = nn.ReLU()
        self.dropout1 = nn.Dropout(0.20)

        self.layer2 = nn.Linear(64, 32)
        self.bn2 = nn.BatchNorm1d(32)
        self.act2 = nn.ReLU()
        self.dropout2 = nn.Dropout(0.20)

        self.layer3 = nn.Linear(32, 16)
        self.bn3 = nn.BatchNorm1d(16)
        self.act3 = nn.ReLU()
        self.dropout3 = nn.Dropout(0.20)

        self.output = nn.Linear(16, 2)  # Adjust the output size based on the number of classes

    def forward(self, x):
        x = self.dropout1(self.act1(self.bn1(self.layer1(x))))
        x = self.dropout2(self.act2(self.bn2(self.layer2(x))))
        x = self.dropout3(self.act3(self.bn3(self.layer3(x))))
        x = self.output(x)
        return x

# class NeuralNet(nn.Module):
#     def __init__(self):
#         super(NeuralNet, self).__init__()
#         # Define the first layer
#         self.layer1 = nn.Linear(128, 64)  # Input from 128 features to 64 features
#         self.bn1 = nn.BatchNorm1d(64)  # Batch normalization for the first layer
#         self.act1 = nn.ReLU()  # Activation function for the first layer
#         self.dropout1 = nn.Dropout(0.30)  # Dropout to prevent overfitting with a 30% drop rate

#         # Define the second layer
#         self.layer2 = nn.Linear(64, 32)  # Reduce from 64 to 32 features
#         self.bn2 = nn.BatchNorm1d(32)  # Batch normalization for the second layer
#         self.act2 = nn.ReLU()  # Activation function for the second layer
#         self.dropout2 = nn.Dropout(0.30)  # Dropout for the second layer

#         # Output layer to produce final predictions
#         self.output = nn.Linear(32, 2)  # Output from 32 features to 2 classes

#     def forward(self, x):
#         # Pass the input through the first layer
#         x = self.dropout1(self.act1(self.bn1(self.layer1(x))))
#         # Pass the output through the second layer
#         x = self.dropout2(self.act2(self.bn2(self.layer2(x))))
#         # Get final output
#         x = self.output(x)
#         return x
```

```python
def evaluate_on_validation_set(model, loader, criterion):
    model.eval()
    total_loss, total_correct, total_samples = 0, 0, 0
    with torch.no_grad():
        for data, target in loader:
            output = model(data)
            loss = criterion(output, target)
            total_loss += loss.item() * data.size(0)
            _, predicted = torch.max(output, 1)
            total_correct += (predicted == target).sum().item()
            total_samples += data.size(0)
    return total_loss / total_samples, total_correct / total_samples * 100
```

```python
def train_client(model, train_loader, val_loader, criterion, optimizer, scheduler, epochs=50, patience=7):
    train_losses, val_losses, train_accuracies, val_accuracies = [], [], [], []
    model.train()
    best_val_loss = float('inf')
    best_val_acc = 0
    patience_counter = 0

    criterion = nn.CrossEntropyLoss(weight=class_weights) # to tacle class imbalance

    for epoch in range(epochs):
        model.train()
        total_loss, total_correct, total_samples = 0, 0, 0

        for data, target in train_loader:
            optimizer.zero_grad()
            output = model(data)
            loss = criterion(output, target)
            loss.backward()
            optimizer.step()
            total_loss += loss.item() * data.size(0)
            _, predicted = torch.max(output, 1)
            total_correct += (predicted == target).sum().item()
            total_samples += data.size(0)

        scheduler.step()  # Update the learning rate

        train_loss = total_loss / total_samples
        train_accuracy = total_correct / total_samples * 100

        # Validation pass
        model.eval()
        val_loss, val_acc = evaluate_on_validation_set(model, val_loader, criterion)

        if val_loss < best_val_loss or (val_loss == best_val_loss and val_acc > best_val_acc):
            best_val_loss = val_loss
            best_val_acc = val_acc
            patience_counter = 0
            torch.save(model.state_dict(), 'best_model.pth')  # Save the best model
        else:
            patience_counter += 1

        if patience_counter >= patience:
            print(f"Stopping early at epoch {epoch+1}")
            break

        print(f'Epoch {epoch+1}: Train Loss: {train_loss:.4f}, Train Accuracy: {train_accuracy:.2f}%, Val Loss: {val_loss:.4f}, Val Accuracy: {val_acc:.2f}%')

    model.load_state_dict(torch.load('best_model.pth'))
    return model.state_dict(), (train_loss, train_accuracy, val_loss, val_acc)
```

```python
# Server-side weight aggregation function
def aggregate_weights(weight_list):
    global_weights = {}
    for key in weight_list[0].keys():
        # Ensure all weights are converted to float, if not already
        temp_weights = [weights[key].float() for weights in weight_list]
        stacked_weights = torch.stack(temp_weights)
        global_weights[key] = stacked_weights.mean(0)
    return global_weights
```

### Training on standardised data

```python
%time

global_model = NeuralNet()
clients_models = [NeuralNet() for _ in range(len(clients_data))]
num_epochs = 20  # Set the number of epochs for training
best_val_accuracy = 0.0  # Track the best validation accuracy
best_model_path = "./models/best_std_global_model.pth"  # Path to save the best model

if not os.path.exists(os.path.dirname(best_model_path)):
    os.makedirs(os.path.dirname(best_model_path))

# Initialize lists to store the average metrics per epoch
avg_train_losses = []
avg_val_losses = []
avg_train_accuracies = []
avg_val_accuracies = []

for epoch in range(num_epochs):
    print(f"Global epoch {epoch + 1}:\n")
    weight_list = []
    train_stats = []
    for i, client_data in enumerate(clients_data):
        print(f"Client data {i}:")
        local_model = clients_models[i]
        local_model.load_state_dict(global_model.state_dict())  # Sync with global model
        optimizer = torch.optim.Adam(local_model.parameters(), lr=0.001, weight_decay=0.001)  # Adding L2 regularization
        scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=10, gamma=0.1)
        criterion = torch.nn.CrossEntropyLoss()
        train_loader, val_loader, _ = client_data
        local_weights, stats = train_client(local_model, train_loader, val_loader, criterion, optimizer, scheduler, epochs=50, patience=7)
        weight_list.append(local_weights)
        train_stats.append(stats)
    global_weights = aggregate_weights(weight_list)
    global_model.load_state_dict(global_weights)

    avg_train_loss = sum(stat[0] for stat in train_stats) / len(train_stats)
    avg_train_accuracy = sum(stat[1] for stat in train_stats) / len(train_stats)
    avg_val_loss = sum(stat[2] for stat in train_stats) / len(train_stats)
    avg_val_accuracy = sum(stat[3] for stat in train_stats) / len(train_stats)
    print(f'Global Epoch {epoch+1}/{num_epochs}: Avg Train Loss: {avg_train_loss:.4f}, Avg Train Accuracy: {avg_train_accuracy:.2f}%, Avg Val Loss: {avg_val_loss:.4f}, Avg Val Accuracy: {avg_val_accuracy:.2f}%')

    # Store the averages for plotting
    avg_train_losses.append(avg_train_loss)
    avg_val_losses.append(avg_val_loss)
    avg_train_accuracies.append(avg_train_accuracy)
    avg_val_accuracies.append(avg_val_accuracy)

    # Check if this is the best model based on validation accuracy
    if avg_val_accuracy > best_val_accuracy:
        best_val_accuracy = avg_val_accuracy
        best_model_state = global_model.state_dict()  # Copy the best state dict
        torch.save(best_model_state, best_model_path)
        print(f"New best model saved with validation accuracy: {best_val_accuracy:.2f}%")

    print()

print(f"Training complete. Best model saved at {best_model_path} with accuracy of {best_val_accuracy:.2f}%.")
```

```python
import numpy as np
from sklearn.metrics import roc_auc_score, classification_report, confusion_matrix, roc_curve
import matplotlib.pyplot as plt
import seaborn as sns

def evaluate_model(model, test_loader, criterion):
    model_path = "./models/best_std_global_model.pth"
    model.load_state_dict(torch.load(model_path))
    model.eval()
    test_loss = 0
    all_predictions = []
    all_probabilities = []  # Store the probabilities for AUC-ROC
    all_targets = []

    with torch.no_grad():
        for data, target in test_loader:
            output = model(data)
            loss = criterion(output, target)
            test_loss += loss.item() * data.size(0)
            _, predicted = torch.max(output, 1)
            probabilities = torch.nn.functional.softmax(output, dim=1)[:, 1]  # Probability of class 1
            all_predictions.extend(predicted.cpu().numpy())
            all_probabilities.extend(probabilities.cpu().numpy())
            all_targets.extend(target.cpu().numpy())

    test_loss /= len(test_loader.dataset)
    accuracy = (np.array(all_predictions) == np.array(all_targets)).mean() * 100
    auc_roc = roc_auc_score(all_targets, all_probabilities)  # Calculate AUC-ROC

    print(f'Test Loss: {test_loss:.4f}, Test Accuracy: {accuracy:.2f}%, AUC-ROC: {auc_roc:.4f}')

    # Calculate classification report and confusion matrix
    print("\nClassification Report:")
    print(classification_report(all_targets, all_predictions, target_names=['Class 0', 'Class 1']))

    # Generate confusion matrix
    cm = confusion_matrix(all_targets, all_predictions)
    print("\nConfusion Matrix:")
    print(cm)

    # Plotting the confusion matrix
    plt.figure(figsize=(8, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Class 0', 'Class 1'], yticklabels=['Class 0', 'Class 1'])
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.title('Confusion Matrix')
    plt.show()

    # Plotting ROC curve
    fpr, tpr, thresholds = roc_curve(all_targets, all_probabilities)
    plt.figure()
    plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % auc_roc)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic')
    plt.legend(loc="lower right")
    plt.show()

    return test_loss, accuracy, auc_roc
```
