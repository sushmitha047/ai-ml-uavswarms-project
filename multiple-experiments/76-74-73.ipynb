```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline

# Calculating the desired counts for each class
desired_attack_count = int(len(complete_data) * 0.6)
desired_natural_count = int(len(complete_data) * 0.4)

# Setting up resampling strategy
over = SMOTE(sampling_strategy={ 'Natural': desired_natural_count })
under = RandomUnderSampler(sampling_strategy={ 'Attack': desired_attack_count })

# Create a pipeline that first oversamples and then undersamples
pipeline = Pipeline([('o', over), ('u', under)])

# Separating features and target variable
X = complete_data.drop('marker', axis=1)
y = complete_data['marker']

# Resampling
X_res, y_res = pipeline.fit_resample(X, y)

# Creating a DataFrame from the resampled data
resampled_df = pd.DataFrame(X_res, columns=X.columns)
resampled_df['marker'] = y_res

# Print new class distribution
new_class_counts = resampled_df['marker'].value_counts()
print("New Class Distribution:")
print(new_class_counts)
```

## Federated learning

```python
import torch
import torch.nn as nn
import numpy as np
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, roc_auc_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

class ClientData(Dataset):
    def __init__(self, features, labels):
        if isinstance(features, pd.Series) or isinstance(features, pd.DataFrame):
            features = features.values
        if isinstance(labels, pd.Series):
            labels = labels.values

        self.features = torch.tensor(features, dtype=torch.float32)
        self.labels = torch.tensor(labels, dtype=torch.float32).unsqueeze(1)  # Ensure labels are the correct shape

    def __len__(self):
        return len(self.features)

    def __getitem__(self, index):
        return self.features[index], self.labels[index]

def prepare_data(df):
    features = df.iloc[:, :-1]
    labels = df['target']
    return features, labels


def train_local_client(model, criterion, optimizer, train_loader, val_loader, num_epochs=50, patience=7):
    best_val_loss = float('inf')
    best_model_state = None
    patience_counter = 0
    
    for epoch in range(num_epochs):
        model.train()
        total_loss, total_samples = 0, 0
        for features, labels in train_loader:
            optimizer.zero_grad()
            outputs = model(features)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            total_loss += loss.item() * len(labels)
            total_samples += len(labels)
        train_loss = total_loss / total_samples
        
        model.eval()
        with torch.no_grad():
            val_loss, val_samples = 0, 0
            for features, labels in val_loader:
                outputs = model(features)
                loss = criterion(outputs, labels)
                val_loss += loss.item() * len(labels)
                val_samples += len(labels)
        val_loss = val_loss / val_samples
        
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            best_model_state = model.state_dict()
            patience_counter = 0
        else:
            patience_counter += 1
            if patience_counter >= patience:
                break
    
    return best_model_state, train_loss, val_loss
```

```python
import torch.nn as nn

class NeuralNet(nn.Module):
    def __init__(self, num_features):
        super(NeuralNet, self).__init__()
        # First layer
        self.layer1 = nn.Linear(num_features, 64)
        self.bn1 = nn.BatchNorm1d(64)
        self.act1 = nn.ReLU()
        self.dropout1 = nn.Dropout(0.2)

        # Second layer
        self.layer2 = nn.Linear(64, 32)
        self.bn2 = nn.BatchNorm1d(32)
        self.act2 = nn.ReLU()
        self.dropout2 = nn.Dropout(0.2)
        
        # Third layer
        self.layer3 = nn.Linear(32, 16)
        self.bn3 = nn.BatchNorm1d(16)
        self.act3 = nn.ReLU()
        self.dropout3 = nn.Dropout(0.2)

        # Output layer
        self.output = nn.Linear(16, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.dropout1(self.act1(self.bn1(self.layer1(x))))
        x = self.dropout2(self.act2(self.bn2(self.layer2(x))))
        x = self.dropout3(self.act3(self.bn3(self.layer3(x))))
        x = self.sigmoid(self.output(x))
        return x
```

```python
def aggregate_weights(models):
    global_weights = {}
    for key in models[0].keys():
        layer_weights = torch.stack([model[key].float() for model in models])
        global_weights[key] = torch.mean(layer_weights, dim=0)
    return global_weights
```

```python
def evaluate_model(model, train_loader, val_loader):
    model.eval()
    train_preds, train_targets, val_preds, val_targets = [], [], [], []
    with torch.no_grad():
        for features, labels in train_loader:
            outputs = model(features).squeeze()
            predicted_labels = (outputs > 0.5).int()
            train_preds.extend(predicted_labels.numpy())
            train_targets.extend(labels.numpy())

        for features, labels in val_loader:
            outputs = model(features).squeeze()
            predicted_labels = (outputs > 0.5).int()
            val_preds.extend(predicted_labels.numpy())
            val_targets.extend(labels.numpy())
    return train_preds, train_targets, val_preds, val_targets

def federated_train(clients_data, num_features, global_epochs=25, local_epochs=40):
    global_model = NeuralNet(num_features)
    global_state_dict = global_model.state_dict()

    global_metrics = {
        'train_accuracy': [], 'val_accuracy': [],
        'train_loss': [], 'val_loss': [],
        'val_recall': [], 'val_precision': [], 'val_f1': []
    }

    best_val_loss = float('inf')
    best_model_state = None

    for global_epoch in range(global_epochs):
        print(f"\nStarting Global Epoch {global_epoch + 1}/{global_epochs}")

        epoch_conf_matrix = np.zeros((2, 2), dtype=int)  # Aggregate confusion matrix for the epoch

        local_metrics = {
            'train_accuracy': [],
            'val_accuracy': [],
            'train_loss': [],
            'val_loss': [],
            'val_recall': [],
            'val_precision': [],
            'val_f1': []
        }

        local_weights_list = []

        for client_index, client_data in enumerate(clients_data, start=1):
            client_model = NeuralNet(num_features)
            client_model.load_state_dict(global_state_dict)

            features, labels = prepare_data(client_data)
            kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
            scaler = StandardScaler()

            for train_idx, val_idx in kfold.split(features, labels):
                train_features = scaler.fit_transform(features.iloc[train_idx])
                val_features = scaler.transform(features.iloc[val_idx])

                train_dataset = ClientData(train_features, labels.iloc[train_idx].values)
                val_dataset = ClientData(val_features, labels.iloc[val_idx].values)
                train_loader = DataLoader(train_dataset, batch_size=128, shuffle=True)
                val_loader = DataLoader(val_dataset, batch_size=128, shuffle=False)

                optimizer = torch.optim.SGD(client_model.parameters(), lr=0.01, momentum=0.9)
                criterion = torch.nn.BCELoss()

                local_state_dict, train_loss, val_loss = train_local_client(client_model, criterion, optimizer, train_loader, val_loader, local_epochs)
                local_weights_list.append(local_state_dict)

                train_preds, train_targets, val_preds, val_targets = evaluate_model(client_model, train_loader, val_loader)
                epoch_conf_matrix += confusion_matrix(val_targets, val_preds)

                local_metrics['train_accuracy'].append(accuracy_score(train_targets, train_preds))
                local_metrics['val_accuracy'].append(accuracy_score(val_targets, val_preds))
                local_metrics['train_loss'].append(train_loss)
                local_metrics['val_loss'].append(val_loss)
                local_metrics['val_recall'].append(recall_score(val_targets, val_preds))
                local_metrics['val_precision'].append(precision_score(val_targets, val_preds))
                local_metrics['val_f1'].append(f1_score(val_targets, val_preds))

            print(f"\nClient Data {client_index}:")
            print(f"Train Accuracy: {local_metrics['train_accuracy'][-1]:.4f}, Val Accuracy: {local_metrics['val_accuracy'][-1]:.4f}")
            print(f"Train Loss: {local_metrics['train_loss'][-1]:.4f}, Val Loss: {local_metrics['val_loss'][-1]:.4f}")

        global_state_dict = aggregate_weights(local_weights_list)
        global_model.load_state_dict(global_state_dict)

        # Update global metrics with the averages of this epoch's metrics
        global_metrics['train_accuracy'].append(np.mean(local_metrics['train_accuracy']))
        global_metrics['val_accuracy'].append(np.mean(local_metrics['val_accuracy']))
        global_metrics['train_loss'].append(np.mean(local_metrics['train_loss']))
        global_metrics['val_loss'].append(np.mean(local_metrics['val_loss']))
        global_metrics['val_recall'].append(np.mean(local_metrics['val_recall']))
        global_metrics['val_precision'].append(np.mean(local_metrics['val_precision']))
        global_metrics['val_f1'].append(np.mean(local_metrics['val_f1']))

        avg_val_loss = np.mean(local_metrics['val_loss'])
        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            best_model_state = global_state_dict.copy()
            print(f"\nNew best model found based on validation loss of: {best_val_loss}")

        print(f"\nAggregated Metrics for Global Epoch {global_epoch + 1}:")
        # print("Average Metrics:", {key: np.mean(value) for key, value in local_metrics.items()})
        print(f"Average Train Accuracy: {np.mean(local_metrics['train_accuracy']):.4f}")
        print(f"Average Val Accuracy: {np.mean(local_metrics['val_accuracy']):.4f}")
        print(f"Average Train Loss: {np.mean(local_metrics['train_loss']):.4f}")
        print(f"Average Val Loss: {np.mean(local_metrics['val_loss']):.4f}")
        print(f"Average Recall: {np.mean(local_metrics['val_recall']):.4f}")
        print(f"Average Precision: {np.mean(local_metrics['val_precision']):.4f}")
        print(f"Average F1 Score: {np.mean(local_metrics['val_f1']):.4f}")

        # Display the aggregated confusion matrix for this epoch
        sns.heatmap(epoch_conf_matrix, annot=True, fmt="d", cmap='Blues')
        plt.title(f'Aggregate Validation Confusion Matrix for Global Epoch {global_epoch + 1}')
        plt.ylabel('True Label')
        plt.xlabel('Predicted Label')
        plt.show()

    global_model.load_state_dict(best_model_state)  # Load the best model
    return global_model, best_model_state, global_metrics, scaler
```

```python
def plot_metrics(global_metrics):
    epochs = range(1, len(global_metrics['train_loss']) + 1)
    plt.figure(figsize=(12, 6))
    
    plt.subplot(1, 2, 1)
    plt.plot(epochs, global_metrics['train_loss'], 'b-', label='Training Loss')
    plt.plot(epochs, global_metrics['val_loss'], 'r-', label='Validation Loss')
    plt.title('Training vs Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(epochs, global_metrics['train_accuracy'], 'b-', label='Training Accuracy')
    plt.plot(epochs, global_metrics['val_accuracy'], 'r-', label='Validation Accuracy')
    plt.title('Training vs Validation Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.tight_layout()
    plt.show()
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

def evaluate_model(model, data_loader):
    model.eval()
    predictions, targets = [], []
    total_loss = 0
    criterion = torch.nn.BCELoss()  # Assuming binary classification

    with torch.no_grad():
        for features, labels in data_loader:
            outputs = model(features)
            loss = criterion(outputs, labels)
            total_loss += loss.item() * len(labels)
            predicted_labels = outputs.round().int()  # Assuming threshold of 0.5 for binary classification
            predictions.extend(predicted_labels.numpy())
            targets.extend(labels.numpy())

    print("Classification Report:")
    print(classification_report(targets, predictions))
    conf_matrix = confusion_matrix(targets, predictions)
    avg_loss = total_loss / len(data_loader.dataset)

    return avg_loss, conf_matrix
```
